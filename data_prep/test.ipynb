```python
# Iterate through the <Sprawozdania> elements
for sprawozdania_elem in root.findall('.//Sprawozdania'):
    for pozycje_elem in sprawozdania_elem.findall('./Pozycje'):
        data_row = {}
        
        # Extract data from <Pozycje>
        for pozycja_elem in pozycje_elem.findall('./Pozycja'):
            for child_elem in pozycja_elem:
                data_row[child_elem.tag] = child_elem.text.strip()
        
        # Extract data from <Sprawozdania>
        for child_elem in sprawozdania_elem:
            data_row[child_elem.tag] = child_elem.text.strip()
        
        rows.append(data_row)

# Create a DataFrame from the extracted data
df = pd.DataFrame(rows)

# Print the DataFrame
df
```
